```python
import matplotlib.pyplot as plt
import numpy as np
from random import sample
%matplotlib inline
```

1. Створіть вектор розміром 11 зі значеннями від 0 до 5. Розрахуйте на
його основі вектор y (y = x^2).

```python
x = np.linspace(0,5,11)
y = x**2
```

2. Побудова графіків.

2.1 Функціональний метод

```python
plt.plot(x,y) # built a plot
plt.xlabel('X Label') # add name of x axis
plt.ylabel('Y Label') # add name of y axis
plt.title('Title') # add title
# plt.show
```

```python
# create multiply plots
plt.subplot(1,2,1) # create first plot: 1 row, 2 colums, 1st plot
plt.plot(x,y,'r')

plt.subplot(1,2,2) # create second plot: 1 row, 2 colums, 2d plot
plt.plot(x,y,'b')
```

2.2 Об'єктно-орієнтований метод

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.1,0.5,0.4,0.3])

axes1.plot(x,y,'r')
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
axes1.set_title('Title')

axes2.plot(x,y,'b')
axes2.set_xlabel('X Label')
axes2.set_ylabel('Y Label')
axes2.set_title('Title')
```

```python
# use similar to plt.figure() except use tuple unpacking to grab fig and axes
fig, axes = plt.subplots()

# now use the axes object to add stuff to plot
axes.plot(x,y,'r')
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Title')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
fig, axes = plt.subplots(nrows=4, ncols=4)
plt.tight_layout()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label = "x**2")
ax.plot(x, x**3, label = "x**3")
ax.legend()
```

```python
fig = plt.figure(figsize = (8,4), dpi = 100)
```

```python
# matlab style line color and style
fig, ax = plt.subplots()
ax.plot(x, x**2, color = 'b') # blue line
ax.plot(x, x**3, color = 'g') # green line
```

```python
# matlab style line color and style
fig, ax = plt.subplots()
ax.plot(x, x**2, lw = 2, ls = '--')
```

```python
# matlab style line color and style
fig, ax = plt.subplots()
ax.plot(x, x**2, marker = 'o', markersize = '10', markerfacecolor = 'y')
```
